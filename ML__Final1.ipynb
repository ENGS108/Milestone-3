Expirementing with the featurization tools in MatMiner 

```python
from pymatgen.core import Composition
from matminer.featurizers.composition import ElementFraction
from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.structure import DensityFeatures
import pandas as pd
import numpy as np


print(type(df_DEC))
print(df_DEC["formula"])
ef = ElementFraction()
comp1 = Composition(df_DEC["formula"][0])
e_frac = ef.featurize(comp1)
print(df_DEC["formula"][0])
print(ef.feature_labels())
print(e_frac)
print(len(e_frac))

stc = StrToComposition()
featured_df = stc.featurize_dataframe(df_DEC, "formula")
print(featured_df.head())
featured_df = ef.featurize_dataframe(featured_df, "composition")
print(featured_df.head())

denf = DensityFeatures()
featured_df = denf.featurize_dataframe(featured_df, "structure")
print(featured_df.head())
print(featured_df.iloc[0])

clean_df = featured_df.drop(["formula", "composition", "structure"], axis=1)
print(len(df_DEC.columns))
print(len(featured_df.columns))
print(len(clean_df.columns))

cleaner_df = clean_df.loc[:, (clean_df != 0).any(axis=0)]
print(len(cleaner_df.columns))

clean_df.to_csv("file.csv",index=False)

```
